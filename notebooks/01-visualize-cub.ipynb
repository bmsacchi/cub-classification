```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/CUB_200_2011/train.csv')
```

```python
df
```

```python
relative_path = lambda x: "../data/processed/CUB_200_2011/images/" + x
```

```python
# Visualize four random images
import matplotlib.pyplot as plt

#sample = df.sample(4)
# filter images that contain the word "warbler" in the filename - for fun!
sample = df[df['filename'].str.contains('Warbler')].sample(4)

fig, axes = plt.subplots(1, 4, figsize=(12, 3))
for ax, (_, row) in zip(axes, sample.iterrows()):
    ax.imshow(plt.imread(relative_path(row['filename'])))

    # Draw bounding box
    x_min, y_min, x_max, y_max = row['x_min'], row['y_min'], row['x_max'], row['y_max']
    ax.add_patch(plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                                 fill=False, edgecolor='red', linewidth=2))
    ax.set_xlabel(f"{row['x_min']}, {row['y_min']}, {row['x_max']}, {row['y_max']}")
    ax.set_xticks([])
    ax.set_yticks([])

    ax.set_title(row['class_name'])
    ax.axis('off')
plt.tight_layout()
```
